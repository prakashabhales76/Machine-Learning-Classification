4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
def Goal_diff_count(data):
    data['Unnamed: 8'] = pd.to_numeric(data['Unnamed: 8'], errors='coerce')
    data['Unnamed: 9'] = pd.to_numeric(data['Unnamed: 9'], errors='coerce')
    data['Goal Difference'] = data['Unnamed: 8'] - data['Unnamed: 9']
    return data[['Unnamed: 1', 'Goal Difference']]
```

```python
goal_diff_df = Goal_diff_count(df)
team_with_max_diff = goal_diff_df.loc[goal_diff_df['Goal Difference'].idxmax()]
team_with_min_diff = goal_diff_df.loc[goal_diff_df['Goal Difference'].idxmin()]
print("Team with maximum goal difference:")
print(team_with_max_diff)
print("\nTeam with minimum goal difference:")
print(team_with_min_diff)
```

5.Create a new column with name “Winning Percent” and append it to the data set (5 points)

```python
def calculate_winning_percent(row):
    games_won = pd.to_numeric(row['Unnamed: 5'], errors='coerce')
    games_played = pd.to_numeric(row['Unnamed: 4'], errors='coerce')
    if not pd.isnull(games_won) and not pd.isnull(games_played) and games_played != 0:
        return (games_won / games_played) * 100
    else:
        return 0
```

```python
df['Winning Percent'] = df.apply(calculate_winning_percent, axis=1)
df.sort_values('Winning Percent', ascending=False, inplace=True)
top_5_teams = df.head(5)[['Unnamed: 1', 'Winning Percent']]
print(top_5_teams)
```
